```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

### Loading the data in Python

```python
df_GDP=pd.read_csv('G:/ProgramData/UpGrad/UpGrad Main Course/5_GDP Assignment/Part 1 A/State-wise_GDP.csv')
df_GDP
```

### Data Cleaning

```python
df_GDP.columns = [c.replace(' ', '_') for c in df_GDP.columns]
df_GDP.columns =[c.strip('_') for c in df_GDP.columns]
```

```python
#Dropping Union Territories
union_ter=["Puducherry","Delhi","Chandigarh","Andaman_&_Nicobar_Islands","Jammu_&_Kashmir"]
```

```python
df_GDP= df_GDP.drop(union_ter,axis = 1)
df_GDP
```

```python
years=pd.DataFrame({'Years':[2012,2013,2014,2015,2016]},index=range(6,11),dtype=int)
df_GDP_2=pd.concat([df_GDP,years],axis=1)
df_GDP_2
```

### Extracting "% Growth over previous year" for all the states

```python
growth_percent_df=df_GDP_2.iloc[6:11,1:]
growth_percent_df
```

```python
#To use maximum data available, we replace the NaN values with mean:
mean_calc=growth_percent_df.mean(axis=0)
mean_calc
```

```python
#Replacing NaNs with mean, to use maximum data available:
growth_percent_df=growth_percent_df.fillna(mean_calc)
growth_percent_df
```

```python
#Now that we have maximum NaN values replaced with mean, we drop the remaining NaN values resulting from lack of data:
growth_percent_df=growth_percent_df.dropna(axis=1)
```

```python
growth_percent_df
```

### Plotting best fit line to represent the growth for each state:

```python
states=[]
states=growth_percent_df.columns
st_list=states[1:-2]
print(st_list)
```

```python
#Plotting % Growth over previous year for all the states:
for i in st_list:
    m,c=np.polyfit(growth_percent_df['Years'],growth_percent_df[i],1)
    plt.figure(figsize=(8,5))
    plt.scatter(x='Years',y=i,data=growth_percent_df)
    plt.plot(growth_percent_df['Years'],m*growth_percent_df['Years']+c,'--r')
    plt.xlabel('Years') 
    plt.ylabel('% Growth over previous year')
    plt.title(i)
    plt.show()
```

Inference:

From the above Graphs it is evident that % growth rates vary for all the states.

Example- Bihar has been almost consistent throughout the years, whereas, Telangana and Goa has shown a consistent and rapid increase in growth rates. However, states like Tamil Nadu and Uttar Pradesh show significant decrease in their rates of growth over the years.

### Plotting best fit line to represent the growth for the nation:

```python
m,c=np.polyfit(growth_percent_df['Years'],growth_percent_df['All_India_GDP'],1)
plt.scatter(x='Years',y='All_India_GDP',data=growth_percent_df)
plt.plot(growth_percent_df['Years'],m*growth_percent_df['Years']+c,'--r')
plt.xlabel('Years') 
plt.ylabel('% Growth over previous year')
plt.title('All_India_GDP')
plt.show()
```

### Plot the percentage contribution of the primary, secondary and tertiary sectors as a percentage of the total GDP for all the states.

Q. Which plot will you use here? Why?

A. A Stacked bar chart with sorted data will give a clear insight into the total GDP contribution of each sector and further the reasons for analysing the calculations. Hence a stacked bar chart is used below.

```python
all_states_GDP_percent=all_states_df[["Population ('00)",'Per Capita GSDP (Rs.)','Primary','Secondary','Tertiary','Taxes on Products','Subsidies on products','TOTAL GSVA at basic prices','Gross State Domestic Product']]
all_states_GDP_percent
```

```python
#Calculating Percentage contribution and adding it into the dataframe:

all_states_GDP_percent['Primary_Sect_Percent']=100*(all_states_GDP_percent['Primary']/all_states_GDP_percent['Gross State Domestic Product'])
all_states_GDP_percent['Secondary_Sect_Percent']=100*(all_states_GDP_percent['Secondary']/all_states_GDP_percent['Gross State Domestic Product'])
all_states_GDP_percent['Tertiary_Sect_Percent']=100*(all_states_GDP_percent['Tertiary']/all_states_GDP_percent['Gross State Domestic Product'])
all_states_GDP_percent['Taxes_Percent']=100*(all_states_GDP_percent['Taxes on Products']/all_states_GDP_percent['Gross State Domestic Product'])
all_states_GDP_percent['Subsidies_Percent']=100*(all_states_GDP_percent['Subsidies on products']/all_states_GDP_percent['Gross State Domestic Product'])
all_states_GDP_percent['GSDP_Percent']=100*(all_states_GDP_percent['Gross State Domestic Product']/all_states_GDP_percent['Gross State Domestic Product'])
all_states_GDP_percent['Total_GSVA_Percent']=100*(all_states_GDP_percent['TOTAL GSVA at basic prices']/all_states_GDP_percent['Gross State Domestic Product'])
all_states_GDP_percent
```

```python
#Plotting a stacked bar chart:

all_states_GDP_percent= all_states_GDP_percent.sort_values('Total_GSVA_Percent') #Sorting the values for a clearer interpretation.
all_states_GDP_percent[['Primary_Sect_Percent', 'Secondary_Sect_Percent', 'Tertiary_Sect_Percent']].plot(kind='bar', grid=True, stacked=True, figsize=(20,10), fontsize=15)
plt.title("Percentage contribution of the total GDP for all the states", fontsize=20)
plt.xlabel("States", fontsize=15)
plt.ylabel("Percentage Contribution", fontsize=15)
```

Q. Why is (Primary + Secondary + Tertiary) not equal to total GDP?

A. Since the taxes and subsidies on the products are excluded from the calculation, the resultant overall contribution of all three sectors does not lead upto 100%.
Since taxes are a part of GDP, and subsidies are subtracted from GDP at factor cost, we must include the calculated difference between taxes and subsidies, to calculate the total GDP.

```python
#To find out why Primary+Secondary+Tertiary is not equal to Total GSDP, we include in the calculation, the Taxes and Subsidies on the product:

all_states_GDP_percent["Including_Tax_Subsid"]= all_states_GDP_percent["Taxes_Percent"]-all_states_GDP_percent["Subsidies_Percent"]
all_states_GDP_percent
```

```python
all_states_GDP_percent[['Primary_Sect_Percent', 'Secondary_Sect_Percent', 'Tertiary_Sect_Percent',"Including_Tax_Subsid"]].plot(kind='bar', grid=True, stacked=True, figsize=(20,10), fontsize=15)
plt.title("Percentage contribution of the total GDP for all the states (including taxes and subsidies)", fontsize=20)
plt.xlabel("States", fontsize=15)
plt.ylabel("Percentage Contribution", fontsize=15)
```

The above graph shows 100% total GDP contribution for all the states, resulting from the contribution of each sector, including the calculated difference between taxes and subsides.
